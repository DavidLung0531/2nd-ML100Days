```python
# -*- coding: utf-8 -*-
"""
Created on Sun May 12 23:53:24 2019

@author: 煥其
"""

import requests
import pandas as pd

target_url = 'https://raw.githubusercontent.com/vashineyu/slides_and_others/master/tutorial/examples/imagenet_urls_examples.txt'

response = requests.get(target_url) #引進資料
data = response.text

# 用 request 傳送回來的資料不會認得斷行符號
print(len(data))

# 找到換行符號，用該符號做字串分割後，把它拿掉
split_tag = '\n'
data = data.split(split_tag)
print(len(data))

name = []
path = []

for i in range(len(data)):
    temp = data[i].split('\t')
    if len(temp) > 1:
        name.append(temp[0])
        path.append(temp[1])
    

arrange_data = {'DID' : name, 'url' : path}

df = pd.DataFrame(arrange_data)
print(df[0:3])



from PIL import Image
from io import BytesIO
import numpy as np
import matplotlib.pyplot as plt

# 請用 df.iloc[...] 得到第一筆資料的連結
first_link = df.iloc[0, 1]
print(first_link)

response = requests.get(first_link)
img = Image.open(BytesIO(response.content))

# Convert img to numpy array

plt.imshow(img)
plt.show()


def img2arr_fromURLs(url_list, resize = False):
    
    """
    請完成這個 Function
    Args
        - url_list: list of URLs
        - resize: bool
    Return
        - list of array
    """
    url_list = range(url_list)
    imglist = []
    for i in url_list: #in 後必須丟可迭代物件 eg. list, range
        link = df.iloc[i, 1]
        print(link)
        response = requests.get(link)
        try:
            img = Image.open(BytesIO(response.content))
            imglist.append(img)
            print('success')
        except:
            print('fail')
    return imglist

result = img2arr_fromURLs(5)
print(result)



print("Total images that we got: %i " % len(result)) # 如果不等於 5, 代表有些連結失效囉

for im_get in result:
    plt.imshow(im_get)
    plt.show()
```
